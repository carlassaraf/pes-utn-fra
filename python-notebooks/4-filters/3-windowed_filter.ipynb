```python
import scipy.signal as sig
import numpy as np
from scipy.fftpack import fft, ifft, fftshift
from __future__ import division
#from ipywidgets import *
import matplotlib.pyplot as plt

numtaps = 100
f = 0.2
b = sig.firwin(numtaps, f)

plt.plot(b)
plt.figure()
f = np.arange(0,1,1/numtaps)
plt.plot(f,abs(fft(b)))

```

Para ver el espectro con la fase vamos a usar: **scipy.signal.freqz**

scipy.signal.freqz(b, a=1, worN=None, whole=0, plot=None)[source]

Compute the frequency response of a digital filter.

Given the numerator b and denominator a of a digital filter, compute its frequency response:

$$H(\omega) = {{B(\omega)} \over {A(\omega)}} =  {{b[0] + b[1] e^{j\omega} + ... + b[m]e^{jm\omega}} \over {a[0] + a[1] e^{j\omega} + ... + a[n]e^{jn\omega}}} $$
     

```python
w, h = sig.freqz(b) # Solo tiene zeros este filtro FIR
fig = plt.figure()
plt.title('Digital filter frequency response')
ax1 = fig.add_subplot(111)

plt.plot(w/(2*np.pi), 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
ax2 = ax1.twinx()

angles = np.unwrap(np.angle(h))
plt.plot(w/(2*np.pi), angles, 'g')
plt.ylabel('Angle (radians)', color='g')
plt.grid()
plt.axis('tight')
plt.show()
```
