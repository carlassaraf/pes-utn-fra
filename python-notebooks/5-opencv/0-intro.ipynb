```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
myPath = ''
img = cv.imread(myPath+'img/mandrill.tiff',1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # Matplotlib espera RGB
plt.xticks([]),plt.yticks([])
plt.imshow(img)

```

```python
# Acceso a pixel
px = img [20,180]
print(px) # R;G;B porque lo convertí arriba.
```

```python
# Acceso a ROI
roi = img[30:80,150:200]
plt.xticks([]),plt.yticks([])
plt.imshow(roi)
```

```python
# Split and merge channels
r,g,b = cv.split(roi)
plt.xticks([]),plt.yticks([])
plt.imshow(b,cmap='gray')
```

```python
g[:,:]=0 # Hace 0 todos los pixeles azules
img_r = cv.merge((r,g,b))
plt.imshow(img_r)
```
