```python
myPath = ''
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

sudo = cv2.imread(myPath+'img/sudoku.jpg',0)
plt.imshow(sudo, cmap='gray')
```

```python
#Scaling
res = cv2.resize(sudo,None,fx=2.5, fy=1.5, interpolation = cv2.INTER_CUBIC)
plt.imshow(res, cmap='gray')
```

```python
# Traslacion
M = np.float32([[1,0,50],[0,1,50]]) # Matriz de desplazamiento
rows,cols = sudo.shape
dst = cv2.warpAffine(sudo,M,(cols,rows))
plt.imshow(dst, cmap='gray')
```

#Matriz de desplazamiento

| 1 0 tx |

| 0 1 ty |

```python
# Rotacion
M = cv2.getRotationMatrix2D((0,0),5,1)
rot = cv2.warpAffine(sudo,M,(cols,rows))
plt.imshow(rot, cmap='gray')
```

```python
# Transformacion de perspectiva
pts1 = np.float32([[55,65],[370,52],[30,383],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2) # Obtiene la matriz de transformacion

dst = cv2.warpPerspective(sudo,M,(300,300))

plt.subplot(121),plt.imshow(sudo, cmap='gray'),plt.title('Input')
plt.subplot(122),plt.imshow(dst, cmap='gray'),plt.title('Output')
plt.show()
```
