```python
myPath = ''
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

lena = cv2.imread(myPath+'img/lena.png')
lenac = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB) # Matplotlib espera RGB
plt.xticks([]), plt.yticks([])
plt.imshow(lenac)
```

```python
K = 15
kernel = np.ones((K,K),np.float32)/(K*K)
dst = cv2.filter2D(lenac,-1,kernel)
plt.subplot(121),plt.imshow(lenac),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv2.blur(lenac,(1,100))
plt.subplot(121),plt.imshow(lenac),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blured')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_noisy = np.copy(lenac)
rows, cols, color = lenac.shape
noise = 100000
for i in range(noise):
        # Pick a random y coordinate
        y_coord = np.random.randint(0, rows - 1)
        # Pick a random x coordinate
        x_coord = np.random.randint(0, cols - 1)
        # Pick a random color
        color_coord = np.random.randint(0,color - 1)
        # Color that pixel to white
        img_noisy[y_coord][x_coord][color_coord] = 255

for i in range(noise):
        # Pick a random y coordinate
        y_coord = np.random.randint(0, rows - 1)
        # Pick a random x coordinate
        x_coord = np.random.randint(0, cols - 1)
        # Pick a random color
        color_coord = np.random.randint(0,color - 1)
        # Color that pixel to white
        img_noisy[y_coord][x_coord][color_coord] = 0


median = cv2.medianBlur(img_noisy,5)

plt.subplot(121),plt.imshow(img_noisy),plt.title('Noisy')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.show()
```
